```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

Написать программу, которая строит график сферы или квадратной пирамиды по выбору пользователя. Выбор фигуры и длины запрашиваются у  пользователя в процессе работы функции


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
figure_name = input('Какая фигура? пирамида или сфера')
if figure_name == 'пирамида':
    h = float(input('Высота: '))
    l = float(input('Длина стороны основания: '))
    steps = 100
    for z, l in zip(np.linspace(0, h, steps), np.linspace(l, 0, steps)):
        x=np.array([l/2, -l/2, -l/2, l/2])
        y=np.array([l/2, l/2, -l/2, -l/2])
        xgrid, ygrid = np.meshgrid(x,y)
        zgrid = xgrid*0 + z
        ax.plot_wireframe(xgrid, ygrid, zgrid,cmap='inferno', alpha=0.3)
elif figure_name == 'сфера':
    r = float(input('Радиус: '))
    u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:100j]
    x = r*np.cos(u)*np.sin(v)
    y = r*np.sin(u)*np.sin(v)
    z = r*np.cos(v)
    ax.plot_surface(x, y, z, cmap='inferno')
    ax.legend()
```

Написать программу (m-файл), которая строит график куба или конуса  по выбору пользователя. Выбор фигуры задается как входные параметры функции.

```python
steps = 100
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
figure_name = input('Какая фигура? куб или конус ')

if figure_name == 'куб':
    l = float(input('Сторона: '))
    for z in zip(np.linspace(0, l, steps)):
        x=np.array([l/2, -l/2, -l/2, l/2])
        y=np.array([l/2, l/2, -l/2, -l/2])
        xgrid, ygrid = np.meshgrid(x,y)
        zgrid = xgrid*0 + z
        ax.plot_wireframe(xgrid, ygrid, zgrid,cmap='inferno', alpha=0.3)
elif figure_name == 'конус':
    r = float(input('Радиус: '))
    h = float(input('Высота: '))
    theta = np.linspace(0, 2 * np.pi, 100)
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    xgrid, ygrid = np.meshgrid(x,y)
    for z, l in zip(np.linspace(0, h, steps), np.linspace(r, 0, steps)):
        x = l * np.cos(theta)
        y = l * np.sin(theta)
        zgrid = xgrid*0 + z
        ax.plot_wireframe(x, y, zgrid, alpha=0.05)

```

Написать программу, которая  создает трехмерную матрицу и  “раскрашивает” её значениями в шахматном порядке по  всему объему. Значение раскрашивания зависят от входного параметра идентификации id:
id = ‘digit’ – матрица заполняется числами 1 и  0;
id = ‘char’ – матрица заполняется буквами ‘1’ и  ‘0’.
Входные параметры – это размерность матрицы по трём координатам,  идентификатор;  выходной – сама матрица.


```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')
m = 4
n = 3

for i in range(3):
    for j in range(3):
        for k in range(3):
            if (i+j+k)%2==1:
                ax.text(2*i, 2*j, 2*k, str(1), color='red')
            else:
                ax.text(2*i, 2*j, 2*k, str(0), color='blue')

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_zlim(-1, 5)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
```

Написать программу, которая создает трёхмерную матрицу,  заполненную  случайными целыми числами (от 0 до 10) и производит арифметические поэлементные действия между двумерными массивами этой матрицы (по любой координате внутри матрицы).  Значения операции зависит от идентификатора:
id = ‘+’
id = ‘–’
id = ‘*’ (поэлементное  умножение).
Размерность матрицы по трём координатам и идентификатор запрашивается у  пользователя в процессе работы функции.  Выходной параметр – полученная матрица.


```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import matplotlib.pyplot as plt
import numpy as np

x = int(input("Размерность по x "))
y = int(input('Размерность по y '))
z = int(input('Размерность по z '))
oper = input('Операция +/-/* ')

fig = plt.figure()
ax = fig.gca(projection='3d')
tenzor1 = np.random.randint(0, 10, size = (x, y, z))
tenzor2 = np.random.randint(0, 10, size = (x, y, z))

if oper == '+':
    tenzor = tenzor1 + tenzor2
elif oper == '-':
    tenzor = tenzor1 - tenzor2
elif oper == '*':
    tenzor = np.multiply(tenzor1, tenzor2)
    
for i in range(x):
    for j in range(y):
        for k in range(z):
            ax.text(i, j, k, str(tenzor[i,j,k]), color='blue')
#             print(i, j, k, tenzor[i,j,k])

ax.set_xlim(0, x)
ax.set_ylim(0, y)
ax.set_zlim(0, z)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
```
